```python
#write a program that asks the user to enter a traffic signal colour,
# based on the input display the appropriate action
colour = input("enter the traffic signal colour:")
if colour == "red":
    print("stop")
elif colour == "yellow":
    print("wait")
elif colour =="green":
    print("go")
else:
    print("invalid colour format")
```

```python
#write a program that asks the use to enter the hour of the day(24 hour format, from 0 to 23)
#print the appropriate greeting based on that(Good morning, good after noon, good evening, good night or invalid hour)

hour = int(input("enter the hour of the day:"))
if hour >= 0 and hour < 12:
    print("good morning")
elif hour >= 12 and hour < 17:
    print("good afternoon")
elif hour >=17 and hour < 21:
    print("good evening")
elif hour >=21 and hour <=23:
    print("good night")
else:
    print("invalid hour")
```

```python
#check the eligibility to vote
# age >= 18, has_id

age = int(input("enter the age:"))
has_id = input("do you have an id (yes/no):")

if age >= 18:
    if has_id == "yes":
        print("you are eligible to vote")

    else:
        print("you do not have an id")
else:
    print("you are not eligible to vote")
```

```python
correct_username = "Abid"
correct_password = "1234"

#take a user input and print login successful, if both the credentials given above are entered, else print invalid name or password
#but the user has only 3 attempts

attempts = 1

while attempts <= 3:
    username = input("enter the username:")
    password = input("enter the password:")
    if username == correct_username and password == correct_password:
        print("login successful")
        break;

    else:
        print("invalid username or password")
        attempts += 1
        if attempts ==3:
            print("you have reached the maximum number of attempts")



```

create a number guessing game. The program should repeatedly ask the user to guess until they get it right. if the user gets it correct then print congratulations, if not print try again. Give the user only 3 attempts to guess.


```python
number = 6
attempts=3
while attempts>0:
 guess=5
 if guess==number:
  print("congratulations")
  break
 else:
    print("try again")
    attempts-=1
    print("you have",attempts,"attempts left")
```
