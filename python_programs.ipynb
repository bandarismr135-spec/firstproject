```python
from datetime import datetime

# --- Greeting and Setup ---
name = input("Enter your name: ").strip()      # string method 1
name = name.title()                             # string method 2
print(f"Hi {name}, welcome to your Daily Expense Assistant!")

# Get today’s date
today = datetime.now().strftime("%Y-%m-%d (%A)")
print(f"Today is {today}")

# Daily spending limit
limit = float(input("Enter your daily spending limit (₹): "))


print("\nWe will now track your expenses for 5 days.\n")

# --- MAIN LOOP FOR 5 DAYS ---
for day in range(1, 6):
    print(f"\nDay {day}")
    print("------------")

    # Expense input
    expense = float(input("Enter today's expense (₹): "))

    # Category input
    category = input("Enter category (food / travel / shopping / other): ").strip().lower()

    # Category-based messages (conditional statements)
    if category == "food":
        print("Healthy spending!")
    elif category == "travel":
        print("Hope it was a safe ride!")
    elif category == "shopping":
        print("Treating yourself is good!")
    else:
        print("Money well spent.")

    # --- Numeric Calculations ---
    tax = expense * 0.10          # 10% tax
    discount = 20                 # fixed discount
    final_amount = expense + tax - discount

    # rounding result
    final_amount = round(final_amount, 2)

    print(f"Final bill after tax & discount: ₹{final_amount}")

    # --- Boolean Check for Limit ---
    within_limit = final_amount <= limit

    if within_limit:
        print("Great! You stayed WITHIN your limit today.")
    else:
        print("ALERT! You EXCEEDED your spending limit today.")

    print(f"--- END OF DAY {day} ---")


print("\nTracking complete! Great job staying consistent, keep going!")
```
